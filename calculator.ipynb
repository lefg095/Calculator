```python
from tkinter import Tk,Text,Button,END,re
from tkinter import *
from tkinter import ttk
import tkinter as tk
from math import *

class Interfaz:
    def __init__(root, ventana):
        root.ventana=ventana
        root.ventana.title("Calculadora")
        root.pantalla=Text(ventana, state="disabled", height=2, width=40, font=("Helvetica",25))
        root.pantalla.grid(row=0, column=0, columnspan=5, rowspan=2, sticky=(W, N, E, S))
        root.operacion=""

        button0 = root.crearBoton(0)
        button1 = root.crearBoton(1)
        button2 = root.crearBoton(2)
        button3 = root.crearBoton(3)
        button4 = root.crearBoton(4)
        button5 = root.crearBoton(5)
        button6 = root.crearBoton(6)
        button7 = root.crearBoton(7)
        button8 = root.crearBoton(8)
        button9 = root.crearBoton(9)

        button_borrar = root.crearBoton(u"\u232B",escribir=False)
        button_borrar_todo = root.crearBoton("C",escribir=False)
        button_parentesis1 = root.crearBoton("(")
        button_parentesis2= root.crearBoton(")")
        button_punto = root.crearBoton(".")

        button_division = root.crearBoton(chr(247))
        button_multiplicacion = root.crearBoton("*")
        button_resta = root.crearBoton("-")
        button_suma = root.crearBoton("+")

        button_igual = root.crearBoton("=",escribir=False)
        button_raiz_cuadrada = root.crearBoton(u"\u221A(")
        button_sin = root.crearBoton("sin(")
        button_cos = root.crearBoton("cos(")
        button_tan = root.crearBoton("tan(")

        #Gestor grid
        button_parentesis1.grid(column=0, row=2, sticky=(W, N, E, S))
        button_parentesis2.grid(column=1, row=2, sticky=(W, N, E, S))
        button_borrar_todo.grid(column=2, row=2, sticky=(W, N, E, S))
        button_borrar.grid(column=3, row=2, columnspan=2, sticky=(W, N, E, S))

        button7.grid(column=0, row=3, sticky=(W, N, E, S))
        button8.grid(column=1, row=3, sticky=(W, N, E, S))
        button9.grid(column=2, row=3, sticky=(W, N, E, S))
        button_division.grid(column=3, row=3, sticky=(W, N, E, S))

        button4.grid(column=0, row=4, sticky=(W, N, E, S))
        button5.grid(column=1, row=4, sticky=(W, N, E, S))
        button6.grid(column=2, row=4, sticky=(W, N, E, S))
        button_multiplicacion.grid(column=3, row=4, sticky=(W, N, E, S))

        button1.grid(column=0, row=5, sticky=(W, N, E, S))
        button2.grid(column=1, row=5, sticky=(W, N, E, S))
        button3.grid(column=2, row=5, sticky=(W, N, E, S))
        button_suma.grid(column=3, row=5, sticky=(W, N, E, S))

        button0.grid(column=0, row=6, columnspan=2,  sticky=(W, N, E, S))
        button_punto.grid(column=2, row=6, sticky=(W, N, E, S))
        button_resta.grid(column=3, row=6, sticky=(W, N, E, S))

        button_igual.grid(column=0, row=7, columnspan=3,  sticky=(W, N, E, S))
        button_raiz_cuadrada.grid(column=3, row=7, sticky=(W, N, E, S))
        button_sin.grid(column=4, row=3, sticky=(W, N, E, S))
        button_cos.grid(column=4, row=4, sticky=(W, N, E, S))
        button_tan.grid(column=4, row=5, sticky=(W, N, E, S))
        
        return

    def crearBoton(root, valor, escribir=True, ancho=9, alto=1):
        return Button(root.ventana, text=valor, width=ancho, height=alto, font=("Helvetica",15),        command=lambda:root.click(valor,escribir))

    def click(root, texto, escribir):
        if not escribir:
            if texto=="=" and root.operacion!="":#calcular operaci√≥n a ser evaluada
                try:
                    root.operacion=re.sub(u"\u00F7", "/", root.operacion)
                    root.operacion=re.sub(u"\u221A", "sqrt", root.operacion)
                    resultado=str(eval(root.operacion))
                    root.operacion=resultado
                    root.limpiarPantalla()
                    root.mostrarEnPantalla(resultado)
                except Exception:
                    root.limpiarPantalla()
                    root.mostrarEnPantalla("Error")
            
            elif texto=="C" or texto=="ErrorC":#limpiar la pantalla
                root.operacion=""
                root.limpiarPantalla()

            elif texto==u"\u232B":#borrar un caracter
                root.operacion = re.sub("sqrt",u"\u221A", root.operacion)
                root.operacion = root.operacion[0:len(root.operacion)-1]
                root.limpiarPantalla()
                root.mostrarEnPantalla(root.operacion)
                return
        else:
            root.operacion+=str(texto)
            root.mostrarEnPantalla(texto)
        return

    def limpiarPantalla(root):
        root.pantalla.configure(state="normal")
        root.pantalla.delete("1.0", END)
        root.pantalla.configure(state="disabled")
        return
    
    def mostrarEnPantalla(root, valor):
        root.pantalla.configure(state="normal")
        root.pantalla.insert(END, valor)
        root.pantalla.configure(state="disabled")
        return


root=tk.Tk()
root.geometry('500x500+700+250')
root.columnconfigure(0, weight=1)
root.columnconfigure(1, weight=1)
root.columnconfigure(2, weight=1)
root.columnconfigure(3, weight=1)
root.columnconfigure(4, weight=1)

root.rowconfigure(0, weight=1)
root.rowconfigure(1, weight=1)
root.rowconfigure(2, weight=1)
root.rowconfigure(3, weight=1)
root.rowconfigure(4, weight=1)
root.rowconfigure(5, weight=1)
root.rowconfigure(6, weight=1)
root.rowconfigure(7, weight=1)
       
calculadora=Interfaz(root)
root.mainloop()
```
